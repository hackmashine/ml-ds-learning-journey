**первая модель ML: линейная регрессия (предсказать цену квартиры).**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.DataFrame({
    "Площадь": [30, 50, 70, 90, 120, 150],
    "Цена": [3, 5, 7, 9, 12, 15]
})

X = data[["Площадь"]]
y = data["Цена"]

model = LinearRegression()
model.fit(X, y)

print("Коэффициент:", model.coef_[0])
print("Свободный член:", model.intercept_)
print("Предсказание для 100м²:", model.predict([[100]])[0])

plt.scatter(X, y, color="blue")
plt.plot(X, model.predict(X), color="red")
plt.xlabel("Площадь (м²)")
plt.ylabel("Цена (млн)")
plt.show()
```
